### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
from heapq import heappop, heappush, heapify
n = int(input('укажите число солбцов')) #число столбцов
m = int(input('укажите число строк')) #число строк
m_min = int(input('укажите мин знач')) #минимальное знач
m_max = int(input('укажите макс знач')) #максимальное значение
matrix = []
def list_creator(n, min, max):#ф-ия создания матрицы
    l = []
    for i in range(n):
        l.append(random.randint(min, max)) #создает списки из случайных чисел в заданном диапозоне
    return l
for i in range(m): #создает матрицу (список в списке)
    matrix.append(list_creator(n, m_min, m_max))
def print_matrix(matrix, m, n): #ф-ия печати м-ы
    for i in range(m):
        for j in range(n):
            print(matrix[i][j], end="   ")
        print('\n')
    print('\n')
print_matrix(matrix, m, n)
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
def sort(list): #сортировка срезами
    for i in range(0, len(list)): #перебираем элементы списка
        clone_list = list[i:]   #создаем клон определенного среза, так как если работать просто со срезом списка, то происходит проблема с индексацией
        list.insert(i, list.pop(clone_list.index(min(clone_list)) + i)) #вставляем на позицию i удаленный по индексу минимальный элемент среза
    return list



start_time = time.time()
for list_1 in matrix:
    list_1 = sort(list_1) #сортируем по срезам
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print_matrix(matrix, m, n)
```

```python
# Сортировка вставкой.
def sort_2(list):
    for i in range(len(list)): #берем элемент списка
        for j in range(i): #проходимся по элементам от начала списка до элемента с индексом i
            if list[i] <= list[j]: #если первоначально взятый элемент меньше или равен элементу, стоящему до него
                list.insert(j, list.pop(i)) #то мы вставляем в позицию j удаленный элемент с позицией i
    return list
print(list)
start_time = time.time()
for list_1 in matrix:
    list_1 = sort_2(list_1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print_matrix(matrix, m, n)
```

```python
# Сортировка обменом. || Сортировка пузырьком.
def sort_p(a):
    for j in range(len(a)):
        for i in range(len(a) - 1):
            if a[i] >= a[i + 1]:
                a[i], a[i + 1] = a[i + 1], a[i]
    return a
start_time = time.time()
for list in matrix:
    list = sort_p(list)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print_matrix(matrix, m, n)
```

```python
# Сортировка Шелла.
def sort(k, matrix):
    for j in range(0, k): #k - интервал Шелла
        for i in range(j, len(matrix), k): #задали шаг, и уже сравниваем элементы с этим шагом
            if i + k <= len(matrix) - 1: #проверка, чтобы сравнение элементов списка не вышло за границы
                if matrix[i + k] < matrix[i]: #если левый  элемент больше правого, 
                    matrix[i], matrix[i + k] = matrix[i + k], matrix[i] #то меняем их местами
    return matrix

def start_sort(list):
    for j in range(len(list) // 2, 0, -2):#здесь мы запускаем сортировку задавая кол-во интервалов относительно тразмнра списка
        for i in range(len(list)): #нужно чтобы сортировка проходила несколько раз с одним интервалом
            list = sort(j, list) 
    for i in range(len(list)):
        list = sort(1, list) #на всякий случай запускаем сортировку с шагом 1 еще пару раз
    return list
start_time = time.time()
for list_1 in matrix:
    list_1 = start_sort(list_1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print_matrix(matrix, m, n)
```

```python
# Быстрая сортировка.

def partition(a, l , h):
    p = random.randint(l, h) #индекс опорного элемента
    a[p], a[h] = a[h], a[p] #меняем местами опорный элемент и последний элемент
    b = l #конечная центральная позиция
    for i in range(l, h): #перебираем весь список
        if a[i] < a[h]:#если текущий элемент меньше последнего(опорного)
            a[i], a[b] = a[b], a[i] #мы меняем местами текущий эл-т и эл-т с центральной позицией
            b += 1 #смещаем цент-ю позицию
    a[h], a[b] = a[b], a[h]
    return b

def quicksort(a, l, h):
    if l < h:
        p = partition(a, l, h)
        quicksort(a, l, p - 1)
        quicksort(a, p + 1, h)
start_time = time.time()
for list_1 in matrix:
    quicksort(list_1, 0, len(list_1) - 1)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print_matrix(matrix, m, n)
```

```python
# Турнирная сортировка.
def merge(array1, array2):
    index1 = index2 = 0
    res = []
    while index1 < len(array1) or index2 < len(array2):
        if index1 >= len(array1):
            res.append(array2[index2])
            index2 += 1
        elif index2 >= len(array2):
            res.append(array1[index1])
            index1 += 1
        elif array1[index1] > array2[index2]:
            res.append(array2[index2])
            index2 += 1
        else:
            res.append(array1[index1])
            index1 += 1
    return res


def tourSort(array):
    heap_size = 15
    length = len(array)
    winners = []
    losers = []

    heap = array[:heap_size] if length > heap_size else array
    heapify(heap)

    index = heap_size
    while index < length:
        current_el = heappop(heap)
        if len(winners) == 0 or winners[-1] <= current_el:
            winners.append(current_el)
        else:
            losers.append(current_el)
        heappush(heap, array[index])
        index += 1

    while heap:
        current_el = heappop(heap)
        if len(winners) == 0 or winners[-1] > current_el:
            winners.append(current_el)
        else:
            losers.append(current_el)

    if len(losers) != 0:
        winners = merge(winners, tourSort(losers))
    return winners

start_time = time.time()
for array in matrix:
    print(tourSort(array))
    print('\n')
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
